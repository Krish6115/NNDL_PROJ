```python
import tensorflow as tf
from tensorflow.keras import layers, models, applications
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import numpy as np
import os
import shutil

# 1. SETUP: CREATE A FOCUSED "APPLE-ONLY" DATASET
# =============================================================
print("--- Setting up a smaller 'Apple-only' dataset ---")

# Define the source directories for Apple classes
original_base_train_dir = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'
original_base_valid_dir = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid'

apple_classes = [
    'Apple___Apple_scab',
    'Apple___Black_rot',
    'Apple___Cedar_apple_rust',
    'Apple___healthy'
]

# Define a new, temporary directory in the writable /kaggle/working/ space
temp_base_dir = '/kaggle/working/apple_dataset'
train_dir = os.path.join(temp_base_dir, 'train')
validation_dir = os.path.join(temp_base_dir, 'valid')

# Create the new directories
os.makedirs(train_dir, exist_ok=True)
os.makedirs(validation_dir, exist_ok=True)

# Create symbolic links for train and validation sets
for class_name in apple_classes:
    # Symlink for training data
    original_path = os.path.join(original_base_train_dir, class_name)
    new_path = os.path.join(train_dir, class_name)
    if not os.path.exists(new_path):
        os.symlink(original_path, new_path)

    # Symlink for validation data
    original_path = os.path.join(original_base_valid_dir, class_name)
    new_path = os.path.join(validation_dir, class_name)
    if not os.path.exists(new_path):
        os.symlink(original_path, new_path)

print(f"Dataset setup complete. Using data from: {temp_base_dir}")
print("--------------------------------------------------")


# 2. DEFINE PATHS AND CONSTANTS
# =============================================================
# Hyperparameters
IMG_SIZE = 224
BATCH_SIZE = 32
LEARNING_RATE = 0.001
EPOCHS = 15

# 3. LOAD AND PREPROCESS DATA
# =============================================================
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    labels='inferred',
    label_mode='int',
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=123
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    validation_dir,
    labels='inferred',
    label_mode='int',
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=False
)

# Get class names and number of classes
class_names = train_ds.class_names
NUM_CLASSES = len(class_names)
print(f"Found {NUM_CLASSES} classes: {class_names}")

# Configure dataset for performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

# Data Augmentation Layer
data_augmentation = models.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
], name="data_augmentation")


# 4. DEFINE THE EFFICIENTNETB0 MODEL
# ==================================
def build_efficientnet_b0(num_classes):
    """
    Builds a model using EfficientNetB0 pre-trained on ImageNet.
    - We freeze the base model to leverage its learned features.
    - We add a new classifier head tailored to our specific task.
    - GlobalAveragePooling2D reduces parameters and helps prevent overfitting.
    """
    base_model = applications.EfficientNetB0(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
    base_model.trainable = False # Freeze the convolutional base

    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = data_augmentation(inputs)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs, outputs, name="EfficientNetB0_Transfer")
    return model

# 5. COMPILE AND TRAIN THE MODEL
# ==================================
# Instantiate the model
efficientnet_model = build_efficientnet_b0(NUM_CLASSES)

# Print summary
print("\n--- EfficientNetB0 Summary ---")
efficientnet_model.summary()

# Compile the model
efficientnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
print("\n--- Training EfficientNetB0 ---")
history = efficientnet_model.fit(
    train_ds,
    epochs=EPOCHS,
    validation_data=val_ds,
    callbacks=[early_stopping]
)

# 6. EVALUATE AND PLOT RESULTS
# ==================================
print("\n--- Model Evaluation on Validation Set ---")
loss, accuracy = efficientnet_model.evaluate(val_ds, verbose=0)
print(f"EfficientNetB0: Accuracy = {accuracy*100:.2f}%")

# Plotting function for a single model
def plot_history(history, model_name):
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    # Plot training & validation accuracy values
    ax1.plot(history.history['accuracy'], linestyle='-', label='Train Accuracy')
    ax1.plot(history.history['val_accuracy'], linestyle='--', label='Validation Accuracy')
    ax1.set_title(f'{model_name} Accuracy')
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.legend(loc='lower right')

    # Plot training & validation loss values
    ax2.plot(history.history['loss'], linestyle='-', label='Train Loss')
    ax2.plot(history.history['val_loss'], linestyle='--', label='Validation Loss')
    ax2.set_title(f'{model_name} Loss')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epoch')
    ax2.legend(loc='upper right')

    plt.suptitle(f'{model_name} Training and Validation Metrics', fontsize=16)
    plt.show()

# Plot the results for your model
plot_history(history, "EfficientNetB0")
```
