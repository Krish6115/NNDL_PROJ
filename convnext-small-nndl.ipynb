```python
# ==============================================================
#  ConvNeXt-Small Baseline Model for Plant Leaf Disease Detection
#  (Pretrained ImageNet, No Fine-Tuning — Fair Baseline)
# ==============================================================

import tensorflow as tf
from tensorflow.keras import layers, models, applications
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
import os, shutil, seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# --------------------------------------------------------------
# Dataset Setup (Apple, Corn, Potato)
# --------------------------------------------------------------

original_train = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'
original_valid = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid'

classes = [
    'Apple___Apple_scab', 'Apple___Black_rot', 'Apple___Cedar_apple_rust', 'Apple___healthy',
    'Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot', 'Corn_(maize)___Common_rust_',
    'Corn_(maize)___Northern_Leaf_Blight', 'Corn_(maize)___healthy',
    'Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy'
]

base = "/kaggle/working/convnext_data"
train_dir = os.path.join(base, "train")
test_dir = os.path.join(base, "test")
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

for c in classes:
    src_t = os.path.join(original_train, c)
    dst_t = os.path.join(train_dir, c)
    if not os.path.exists(dst_t): shutil.copytree(src_t, dst_t)

    src_v = os.path.join(original_valid, c)
    dst_v = os.path.join(test_dir, c)
    if not os.path.exists(dst_v): shutil.copytree(src_v, dst_v)

print("✅ Data dirs ready.")

# --------------------------------------------------------------
# Training Hyperparameters
# --------------------------------------------------------------

IMG_SIZE = 224
BATCH = 32
EPOCHS = 20
LR = 1e-3
VAL_SPLIT = 0.2
SEED = 42
DROPOUT = 0.5

# --------------------------------------------------------------
# Load and Split Dataset
# --------------------------------------------------------------

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir, image_size=(IMG_SIZE,IMG_SIZE), batch_size=BATCH,
    shuffle=True, seed=SEED, validation_split=VAL_SPLIT, subset="training"
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir, image_size=(IMG_SIZE,IMG_SIZE), batch_size=BATCH,
    shuffle=False, seed=SEED, validation_split=VAL_SPLIT, subset="validation"
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir, image_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH, shuffle=False
)

class_names = train_ds.class_names
CLASSES = len(class_names)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)
test_ds = test_ds.prefetch(tf.data.AUTOTUNE)

augment = models.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1)
])

# --------------------------------------------------------------
# Build ConvNeXt Model (Baseline)
# --------------------------------------------------------------

def build_convnext(num_classes):
    base = applications.ConvNeXtSmall(
        include_top=False,
        weights="imagenet",
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )
    base.trainable = False

    inp = layers.Input(shape=(IMG_SIZE,IMG_SIZE,3))
    x = augment(inp)
    x = applications.convnext.preprocess_input(x)
    x = base(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(DROPOUT)(x)
    out = layers.Dense(num_classes, activation="softmax")(x)

    return models.Model(inp, out)

model = build_convnext(CLASSES)
model.summary()

# --------------------------------------------------------------
# Compile & Train
# --------------------------------------------------------------

model.compile(optimizer=tf.keras.optimizers.Adam(LR),
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

cb = [
    EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True),
    ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=2, min_lr=1e-6)
]

history = model.fit(train_ds, epochs=EPOCHS, validation_data=val_ds, callbacks=cb)

# --------------------------------------------------------------
# Plot Curves
# --------------------------------------------------------------

def show_history(hist,name):
    a,v = hist.history['accuracy'], hist.history['val_accuracy']
    l,vl = hist.history['loss'], hist.history['val_loss']
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1); plt.plot(a); plt.plot(v); plt.title(f'{name} Accuracy')
    plt.subplot(1,2,2); plt.plot(l); plt.plot(vl); plt.title(f'{name} Loss')
    plt.show()

show_history(history, "ConvNeXt-Small")

# --------------------------------------------------------------
# Evaluation
# --------------------------------------------------------------

loss,acc = model.evaluate(test_ds)
print(f"✅ ConvNeXt-Small Test Accuracy: {acc*100:.2f}%")
print(f"✅ ConvNeXt-Small Test Loss: {loss:.4f}")

# --------------------------------------------------------------
# Classification Report & Confusion Matrix
# --------------------------------------------------------------

y_true = np.concatenate([y for _,y in test_ds])
y_prob = model.predict(test_ds)
y_pred = np.argmax(y_prob, axis=1)

print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=class_names, yticklabels=class_names, cmap="Blues")
plt.title("ConvNeXt-Small Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.show()
```
