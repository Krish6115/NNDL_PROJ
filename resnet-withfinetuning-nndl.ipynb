```python
import tensorflow as tf
from tensorflow.keras import layers, models, applications
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
import os
import shutil  # Make sure shutil is imported
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# 1. SETUP: CREATE FOCUSED "MULTI-PLANT" DATASET
# ==============================================================================
print("--- Setting up 'Apple, Corn, Potato' train, validation, and test datasets ---")

# --- Part A: Define all 11 classes ---
original_base_train_dir = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'
original_base_valid_dir = '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid' # This will be our test set

all_classes = [
    'Apple___Apple_scab', 'Apple___Black_rot', 'Apple___Cedar_apple_rust', 'Apple___healthy',
    'Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot', 'Corn_(maize)___Common_rust_',
    'Corn_(maize)___Northern_Leaf_Blight', 'Corn_(maize)___healthy',
    'Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy'
]
print(f"Total classes to load: {len(all_classes)}")

# --- Part B: Setup Train, Validation, and Test Dirs (using copytree) ---
temp_base_dir = '/kaggle/working/endsem_project_dataset'
train_dir = os.path.join(temp_base_dir, 'train')
test_dir = os.path.join(temp_base_dir, 'test') # We'll copy the 'valid' data here

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

print("Copying Train and Test (from Valid) files... This may take 1-2 minutes.")

for class_name in all_classes:
    # --- Train Data ---
    original_path = os.path.join(original_base_train_dir, class_name)
    new_path = os.path.join(train_dir, class_name)
    if not os.path.exists(new_path) and os.path.exists(original_path):
        shutil.copytree(original_path, new_path) # Use copytree

    # --- Test Data (from original_base_valid_dir) ---
    original_path = os.path.join(original_base_valid_dir, class_name)
    new_path = os.path.join(test_dir, class_name)
    if not os.path.exists(new_path) and os.path.exists(original_path):
        shutil.copytree(original_path, new_path) # Use copytree

print("Train and Test directories created via copy.")
print("----------------------------------------------------------------------")


# 2. DEFINE CONSTANTS AND HYPERPARAMETERS
# ==============================================================================
IMG_SIZE = 224
BATCH_SIZE = 32
VALIDATION_SPLIT = 0.20  # Use 20% of the training data for validation
RANDOM_SEED = 123

# Training Hyperparameters
DROPOUT_RATE = 0.5
INITIAL_EPOCHS = 10
FINE_TUNE_EPOCHS = 10
TOTAL_EPOCHS = INITIAL_EPOCHS + FINE_TUNE_EPOCHS
INITIAL_LR = 0.001
FINE_TUNE_LR = 1e-5

# 3. LOAD AND PREPROCESS DATA
# ==============================================================================
print("Loading and splitting dataset...")

# Create Training (80%) and Validation (20%) sets from the 'train_dir'
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=RANDOM_SEED,
    validation_split=VALIDATION_SPLIT,
    subset="training"
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=False,
    seed=RANDOM_SEED,
    validation_split=VALIDATION_SPLIT,
    subset="validation"
)

# Create the final Test set from the separate 'test_dir'
test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
NUM_CLASSES = len(class_names)
print(f"Found {NUM_CLASSES} classes: {class_names}")

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE) # Cache the test set

# Data Augmentation Layer
data_augmentation = models.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="data_augmentation")


# 4. DEFINE THE RESNET50V2 MODEL (Transfer Learning)
# ==============================================================================
def build_resnet_model(num_classes):
    base_model = applications.ResNet50V2(
        include_top=False,
        weights='imagenet',
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        name="resnet50v2_base"
    )
    base_model.trainable = False
    
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = data_augmentation(inputs)
    x = applications.resnet_v2.preprocess_input(x)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(DROPOUT_RATE)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)
    
    model = models.Model(inputs, outputs, name="ResNet50V2_Transfer")
    return model

# 5. COMPILE AND TRAIN (TWO-STAGE PROCESS)
# ==============================================================================
resnet_model = build_resnet_model(NUM_CLASSES)
print("\n--- ResNet50V2 Summary (Base Frozen) ---")
resnet_model.summary()

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6, verbose=1)

# --- Stage 1: Train the Classifier Head ---
resnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=INITIAL_LR),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
print("\n--- Stage 1: Training Head (ResNet Base Frozen) ---")
history = resnet_model.fit(
    train_ds,
    epochs=INITIAL_EPOCHS,
    validation_data=val_ds, # Validate on the 20% split
    callbacks=[early_stopping, lr_scheduler]
)

# --- Stage 2: Fine-Tune the Base Model ---
resnet_model.get_layer("resnet50v2_base").trainable = True
for layer in resnet_model.get_layer("resnet50v2_base").layers[:-30]:
    layer.trainable = False

resnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=FINE_TUNE_LR),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
print("\n--- Stage 2: Fine-Tuning ResNet Base ---")
history_fine = resnet_model.fit(
    train_ds,
    epochs=TOTAL_EPOCHS,
    initial_epoch=history.epoch[-1],
    validation_data=val_ds, # Validate on the 20% split
    callbacks=[early_stopping]
)

# 6. PLOT TRAINING & VALIDATION HISTORY
# ==============================================================================
def plot_fine_tune_history(history, history_fine):
    acc = history.history['accuracy'] + history_fine.history['accuracy']
    val_acc = history.history['val_accuracy'] + history_fine.history['val_accuracy']
    loss = history.history['loss'] + history_fine.history['loss']
    val_loss = history.history['val_loss'] + history_fine.history['val_loss']

    plt.style.use('seaborn-v0_8-whitegrid')
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

    ax1.plot(acc, 'o-', label='Train Accuracy')
    ax1.plot(val_acc, 'o--', label='Validation Accuracy')
    ax1.axvline(x=len(history.history['accuracy'])-1, color='gray', linestyle='--', label='Start of Fine-Tuning')
    ax1.set_title('Model Accuracy')
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.legend()

    ax2.plot(loss, 'o-', label='Train Loss')
    ax2.plot(val_loss, 'o--', label='Validation Loss')
    ax2.axvline(x=len(history.history['loss'])-1, color='gray', linestyle='--', label='Start of Fine-Tuning')
    ax2.set_title('Model Loss')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epoch')
    ax2.legend()
    plt.suptitle('ResNet50V2 Fine-Tuning Performance', fontsize=16)
    plt.show()

plot_fine_tune_history(history, history_fine)


# 7. FINAL EVALUATION AND CLASSIFICATION METRICS (on the REAL test set)
# ==============================================================================
print("\n--- Model Evaluation on Test Set (from 'valid' folder) ---")

# We evaluate on test_ds, which is our real test set
loss, accuracy = resnet_model.evaluate(test_ds, verbose=1)
print(f"✅ ResNet50V2 Test Accuracy: {accuracy*100:.2f}%")
print(f"✅ ResNet50V2 Test Loss: {loss:.4f}")

print("\n--- Generating Classification Report and Confusion Matrix ---")
y_true = np.concatenate([y for x, y in test_ds], axis=0)
y_pred_probs = resnet_model.predict(test_ds)
y_pred = np.argmax(y_pred_probs, axis=1)

# === THE FIX FOR THE VALUEERROR IS HERE ===
# We explicitly tell the report what all the labels are.
labels = np.arange(len(class_names))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, labels=labels, zero_division=0))

print("\nConfusion Matrix:")
cm = confusion_matrix(y_true, y_pred, labels=labels) # Use the same labels
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix for ResNet50V2 (on Test Set)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
